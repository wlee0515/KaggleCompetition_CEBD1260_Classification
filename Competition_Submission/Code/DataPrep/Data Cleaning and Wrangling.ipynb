### Name

```python
list_Name = df_train["Name"].unique()
print(list_Name)
list_Name = df_test["Name"].unique()
print(list_Name)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

list_name_column = [df_train_new["Name"], df_test_new["Name"]]
for i in range(0,2):
    list_name_column[i] = list_name_column[i].fillna("No Name")
    list_name_column[i] = list_name_column[i].str.replace(r"[^A-Za-z]", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"^ ", "")
    list_name_column[i] = list_name_column[i].str.replace(r" $", "")
    list_name_column[i] = list_name_column[i].apply(lambda x: "No Name" if (x=="") else stemmer.stem(x))
    
df_train_new["Name"] = list_name_column[0]
df_test_new["Name"] = list_name_column[1]
```

```python
total_name_count = list_name_column[0].count() + list_name_column[1].count()
total_name_list = pd.concat([list_name_column[0],list_name_column[1]], axis=0)
total_name_list = total_name_list.value_counts().reset_index()
total_name_list.columns = ["Name", "Count"]
total_name_list["Probability"] = total_name_list["Count"].apply(lambda x : x / total_name_count)
total_name_list
```

```python
total_name_list = total_name_list.drop("Count", axis=1)

df_train_new = df_train_new.merge(total_name_list,on="Name", how="left")
df_train_new = df_train_new.drop("Name", axis=1)
df_test_new = df_test_new.merge(total_name_list,on="Name", how="left")
df_test_new = df_test_new.drop("Name", axis=1)
```

### Breed

```python
list_breed = [] 
list_breed.extend(df_train["Breed1"].unique().tolist())
list_breed.extend(df_train["Breed2"].unique().tolist())
list_breed.extend(df_test["Breed1"].unique().tolist())
list_breed.extend(df_test["Breed2"].unique().tolist())
list_breed = list(set(list_breed))
print(list_breed)
```

```python
for i in list_breed:
    df_train_new["Breed_{0}".format(i)] = 0
    df_test_new["Breed_{0}".format(i)] = 0

for iter, row in df_train_new.iterrows():
    row["Breed_{0}".format(row["Breed1"])] += 1
    row["Breed_{0}".format(row["Breed2"])] += 1

for iter, row in df_test_new.iterrows():
    row["Breed_{0}".format(row["Breed1"])] += 1
    row["Breed_{0}".format(row["Breed2"])] += 1
```

```python
df_train_new = df_train_new.drop("Breed1", axis=1)
df_train_new = df_train_new.drop("Breed2", axis=1)
df_test_new = df_test_new.drop("Breed1", axis=1)
df_test_new = df_test_new.drop("Breed2", axis=1)
```

### Color

```python
list_Color = [] 
list_Color.extend(df_train["Color1"].unique().tolist())
list_Color.extend(df_train["Color2"].unique().tolist())
list_Color.extend(df_train["Color3"].unique().tolist())
list_Color.extend(df_test["Color1"].unique().tolist())
list_Color.extend(df_test["Color2"].unique().tolist())
list_Color.extend(df_test["Color3"].unique().tolist())
list_Color = list(set(list_Color))
print(list_Color)
```

```python
df_train_new["Color1"] = df_train_new["Color1"].fillna(0)
df_train_new["Color2"] = df_train_new["Color2"].fillna(0)
df_train_new["Color3"] = df_train_new["Color3"].fillna(0)
df_test_new["Color1"] = df_test_new["Color1"].fillna(0)
df_test_new["Color2"] = df_test_new["Color2"].fillna(0)
df_test_new["Color3"] = df_test_new["Color3"].fillna(0)


for i in list_Color:
    df_train_new["Color_{:d}".format(i)] = 0
    df_test_new["Color_{:d}".format(i)] = 0

for iter, row in df_train_new.iterrows():
    row["Color_{:d}".format(int(row["Color1"]))] += 1
    row["Color_{:d}".format(int(row["Color2"]))] += 1
    row["Color_{:d}".format(int(row["Color3"]))] += 1

for iter, row in df_test_new.iterrows():
    row["Color_{:d}".format(int(row["Color1"]))] += 1
    row["Color_{:d}".format(int(row["Color2"]))] += 1
    row["Color_{:d}".format(int(row["Color3"]))] += 1
```

```python
df_train_new = df_train_new.drop("Color1", axis=1)
df_train_new = df_train_new.drop("Color2", axis=1)
df_train_new = df_train_new.drop("Color3", axis=1)
df_test_new = df_test_new.drop("Color1", axis=1)
df_test_new = df_test_new.drop("Color2", axis=1)
df_test_new = df_test_new.drop("Color3", axis=1)
```
