### Name

```python
list_Name = df_train["Name"].unique()
print(list_Name)
list_Name = df_test["Name"].unique()
print(list_Name)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

list_name_column = [df_train_new["Name"], df_test_new["Name"]]
for i in range(0,2):
    list_name_column[i] = list_name_column[i].str.lower()
    list_name_column[i] = list_name_column[i].fillna("no_name")
    list_name_column[i] = list_name_column[i].str.replace(r"[^A-Za-z]", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"  ", " ")
    list_name_column[i] = list_name_column[i].str.replace(r"^ ", "")
    list_name_column[i] = list_name_column[i].str.replace(r" $", "")
    list_name_column[i][(list_name_column[i] == "")] = "no_name"
    list_name_column[i][(list_name_column[i] == "No Name")] = "no_name"
    list_name_column[i] = list_name_column[i].apply(lambda x : "no_name" if ((x == "no_name") | (-1 != x.find("no name"))) else x )
    list_name_column[i] = list_name_column[i].apply(lambda x : "phrase_name" if ((x != "no_name") & (len(x.split(" ")) >=3)) else x )
        
df_train_new["Name"] = list_name_column[0]
df_test_new["Name"] = list_name_column[1]
```

```python
total_name_count = list_name_column[0].count() + list_name_column[1].count()
total_name_list = pd.concat([list_name_column[0],list_name_column[1]], axis=0)
total_name_list = total_name_list.value_counts().reset_index()
total_name_list.columns = ["Name", "Count"]
total_name_list["Probability"] = total_name_list["Count"].apply(lambda x : 100*x / total_name_count)
total_name_list
```

```python
encode_list_1 = list_name_column[0].tolist()
encode_list_2 = list_name_column[1].tolist()
encode_list_Total = total_name_list["Name"].tolist()

encode_list = [ name for name in encode_list_Total if ((name in encode_list_1)&(name in encode_list_2))]

for name in encode_list:
    df_train_new["Name_{0}".format(name)] = 0
    df_test_new["Name_{0}".format(name)] = 0

for index, row in df_train_new.iterrows():
    name_str = row["Name"]
    if name_str in encode_list:
        row["Name_{0}".format(name_str)] = 1

for index, row in df_test_new.iterrows():
    name_str = row["Name"]
    if name_str in encode_list:
        row["Name_{0}".format(name_str)] = 1
        
```

```python
df_train_new["has_a_name"] = df_train_new["Name"].apply(lambda x: 0 if x == "no_name" else 1)
df_test_new["has_a_name"] = df_train_new["Name"].apply(lambda x: 0 if x == "no_name" else 1)
```

```python
total_name_list = total_name_list.drop("Count", axis=1)

df_train_new = df_train_new.merge(total_name_list,on="Name", how="left")
df_train_new = df_train_new.drop("Name", axis=1)

df_test_new = df_test_new.merge(total_name_list,on="Name", how="left")
df_test_new = df_test_new.drop("Name", axis=1)
```

### Color

```python
list_Color = [] 
list_Color.extend(df_train["Color1"].unique().tolist())
list_Color.extend(df_train["Color2"].unique().tolist())
list_Color.extend(df_train["Color3"].unique().tolist())
list_Color.extend(df_test["Color1"].unique().tolist())
list_Color.extend(df_test["Color2"].unique().tolist())
list_Color.extend(df_test["Color3"].unique().tolist())
list_Color = list(set(list_Color))
print(list_Color)
```
